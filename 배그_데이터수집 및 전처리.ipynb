## 데이터프레임 만들기

* 'type': 'participant 

    => 64명의 선수 정보가 담겨있는 딕셔너리
* 'type': 'roster'

    => rank, teamId 등 정보가 담겨있는 딕셔너리

```python
# 64명 선수 정보만 데이터 프레임으로 저장하기
participant_indexes = []
for i, item in enumerate(mat_json['included']):
    if item.get('type') == 'participant':
        participant_indexes.append(i)

dfs = []
for i in participant_indexes:
    mat_player = mat_json['included'][i]['attributes']['stats']
    df_mat_player = pd.DataFrame(mat_player, index=[0])
    dfs.append(df_mat_player)

result_df = pd.concat(dfs, ignore_index=True)
result_df
```

```python
from pubg_python import PUBG, Shard
from pubg_python.domain.telemetry.events import LogMatchEnd
```

```python
api = PUBG(api_key=key, shard=Shard.PC_KRJP)

# 프로 선수의 닉네임을 지정합니다.
player_name = player

# 매치 정보를 가져오기 위한 필터를 지정합니다.
filter = {'playerNames': [player_name]}

# 매치 리스트를 가져옵니다.
matches = api.matches(filter=filter)

# 최근 5개의 매치에서 매치 정보를 출력합니다.
for match in matches[:5]:
    # 매치 정보를 가져옵니다.
    telemetry = api.telemetry(match.id)

    # 매치 정보에서 매치 종료 이벤트를 찾습니다.
    match_end_event = telemetry.find_first(LogMatchEnd)

    # 매치 종료 이벤트에서 우승자 정보를 가져옵니다.
    winner = match_end_event.winners[0]

    print(f'Match ID: {match.id}, Map: {match.map_name}, Mode: {match.game_mode}, Winner: {winner.name}')
```
