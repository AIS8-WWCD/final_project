## (3)match id로 64명 player 정보 가져오기

```python
match_id = df_mat['id'][0]
match_id
```

```python
url_mat = f'https://api.pubg.com/shards/tournaments/matches/{match_id}'
r_mat = requests.get(url_mat, headers=header)
mat_json = r_mat.json()
mat_json
```

* 'type': 'participant 

    => 64명의 선수 정보가 담겨있는 딕셔너리
* 'type': 'roster'

    => rank, teamId 등 정보가 담겨있는 딕셔너리

```python
# 64명 선수 정보만 데이터 프레임으로 저장하기
participant_indexes = []
for i, item in enumerate(mat_json['included']):
    if item.get('type') == 'participant':
        participant_indexes.append(i)

dfs = []
for i in participant_indexes:
    mat_player = mat_json['included'][i]['attributes']['stats']
    df_mat_player = pd.DataFrame(mat_player, index=[0])
    dfs.append(df_mat_player)

df_match1 = pd.concat(dfs, ignore_index=True)
df_match1 = df_match1.sort_values(by='winPlace')
df_match1
```

```python
# 팀명 칼럼 생성하기 (name에서 '_' 앞에 해당하는 부분)
df_match1['team'] = df_match1['name'].map(lambda x: x.split('_')[0])
# 팀 개수 16개 맞는지 확인
print(len(df_match1['team'].unique()))
# 팀 고유값 확인
print(df_match1['team'].unique())
# 최종 데이터프레임 확인
display(df_match1)
```

## (4)Telemetry url
* 경기 세부 로그 불러오는 url 찾기

```python
url_list = []

for i in range(18):
    match_id = df_mat['id'][i]
    url_tele = f'https://api.pubg.com/shards/tournaments/matches/{match_id}'
    rs_tele = requests.get(url_tele, headers=header)
    tele_json = rs_tele.json()
    # url 찾아내자.
    url_i = 0
    for i, asset in enumerate(tele_json['included']):
        if asset.get('type') == 'asset':
            url_i = url_i +i
    telemetry_url = tele_json['included'][url_i]['attributes']['URL']
    url_list.append(telemetry_url)

url_list
```

```python
# df_mat에 새로운 칼럼으로 넣기
df_mat['telemetry_url'] = pd.DataFrame(url_list)
df_mat
```

2. 선수별로 적중률을 구하기
* 총 발사한 총알 수
* 데미지를 입힌 총알 수
의 정보를 알려주는 로그를 찾기

```python
# LogPlayerAttack 과 LogPlayerTakeDamage 를 활용하기
# NAVI 팀을 기준으로 샘플링 해봄
df_attack_log = df_log[(df_log['_T']=='LogPlayerAttack') | (df_log['_T']=='LogPlayerTakeDamage')]
df_attack_log = df_attack_log[df_attack_log['attacker'].astype(str).str.contains('NAVI')].reset_index(drop=True)
df_attack_log
```

```python
# 없으면 실행이안됨, 'attacker_name' 컬럼을 만들어주기 위함
df_attack_log['attacker_name'] = np.NaN 

# 데이터프레임 칼럼 안에 넣기
for i in range(len(df_attack_log)):
    df_attack_log['attacker_name'][i] = df_attack_log[df_attack_log['attacker'].astype(str).str.contains('NAVI')]['attacker'][i]['name']

# NAVI팀 선수들의 명중률    
NAVI_attack_log = df_attack_log.groupby('attacker_name').count()
team_hit_data = pd.DataFrame(NAVI_attack_log['damage']/NAVI_attack_log['fireWeaponStackCount'] * 100)
team_hit_data = team_hit_data.unstack().reset_index()
team_hit_data = team_hit_data[['attacker_name',0]]
team_hit_data = team_hit_data.rename(columns={0:'hit_rate'})
team_hit_data
```
